Por definicin un VCS (*Version Control System*) es una herramienta que permite llevar un control del historial de cambios y versiones de un conjunto de archivos de manera organizada y con la capacidad incluso de almacenar la fecha  y el usuario que realiz las modificaciones. Permite trabajar en varias versiones del mismo archivo por diferentes usuarios y luego unificar los cambios para formar un archivo principal; tambin es capaz de recuperar una versin anterior y restablecerla como archivo principal; la analoga utilizada es de un rbol y sus ramas, representando el rbol al conjunto de archivos, y las ramas a cada una de las variantes de los archivos que lo componen. 
En el rea del desarrollo de software se ha convertido en un **estndar** contar con un VCS, incluso algunos de los *Frameworks* o libreras de reciente creacin lo incluyen por defecto al iniciar un proyecto bajo su esquema.

Existen varios VCS, sin embargo GIT se ha convertido en la principal y ms utilizada opcin en el rea del desarrollo de software. Es un sistema de licencia libre creado por Linus Torvalds y que ha sido adoptado como el principal sistema de control de versiones en el mundo. 

Algunos de los trminos bsicos de un GIT:
* **Repositorio:** Es el nombre que recibe el conjunto de archivos que conforman el proyecto.
* **Rama:** Son las bifurcaciones que se hacen a un archivo para poder trabajar en nuevas caractersticas y que 
    posteriormente pueden ser incorporadas al archivo principal
* **Commit:** Es la accin de confirmar y guardar los cambios que se han hecho a uno o varios archivos
* **Merge:** Es la accin de incorporar dos diferentes ramas para unificar cambios y formar un solo archivo 
    actualizado
* **Feature:** Es el nombre en ingls para una caracterstica, nominalmente es el nombre que reciben las funcionalidades que se trabajan en una rama
    
## GIT para desarrollo de software
El modelo de datos de GIT se basa en la generacin de snapshots para cada una de las versiones de una coleccin de archivos que almacena los datos necesarios para llevar a cabo el control de manera organizada adems de un apuntador hacia su origen. El rbol principal y  sus diferentes ramificaciones se identifican con un bash nico que permite localizarlos en un universo de commits.  Bsicamente un rbol genera ramas, las ramas son modificadas y estas pueden ser unificadas para formar un nuevo archivo principal. Enfocados en el desarrollo de software, la principal utilidad de este mecanismo es el permitir trabajar en una nueva funcionalidad en un ambiente totalmente aislado de la rama principal y, luego de las pruebas pertinentes, ser incorporado a la rama principal. Graficamente se puede expresar de la siguiente manera:

````
o <-- o <-- o <-- o <---- o
            ^            /
             \          v
              --- o <-- o
````
* o: Rama
* <--: Apuntador

### rea de staging
Antes de confirmar un cambio y guardarlo definitivamente en el repositorio GIT nos provee de un rea de staging, la cual es un lugar donde podemos revisar los diferentes cambios que hemos hecho a los diferentes archivos previo a crear un commit. Todos los cambios que se hacen a una rama no necesariamente deben guardarse en un mismo commit, podemos segmentarlos de acuerdo a la necesidad, o simplemente para explicar mejor el trabajo realizado. Es un rea muy til una vez el proyecto crece en cantidad de archivos y usuarios desarrolladores.

### Lnea de comandos de GIT
GIT es una herramienta que naci para ser utilizada por medio de la lnea de comandos, por lo tanto al instalarla en cualquier sistema operativo es esa la puerta de acceso principal y la ms recomendada debido a la libertad nativa que ofrece, sin embargo, existen varias interfaces grficas para utilizarlo y que ofrecen un apoyo al momento de visualizar las diferentes ramas del repositorio, los autores, fechas de actualizacin y mensajes del commit realizado. 

GIT ofrece un grupo de comandos para poder interactuar con la herramienta y llevar a cabo nuestras tareas, el listado es amplio pero son los comandos esenciales son los siguientes: 

* git help <command>: Provee ayuda para un comando
* git init: Inicializa un repositorio en el directorio donde se ejecuta
* git status: Le indica al usuario el estado de la rama o repositorio actual
* git add < filename>: Agrega un archivo al rea de staging
* git commit: Crea un nuevo commit con los archivos en el rea de staging, se le pude agregar un mensaje al concatenar la bandera -m "Message"
* git log: Muestra un historial
* git checkout <branch | revision>: Permite moverse entre ramas
* git branch: Muestra un listado de las ramas que se tienen localmente
* git branch < name>: Crea una nueva rama
* git checkout -b < name>: Crea una nueva rama y nos mueve hacia ese espacio
* git merge <branch | revision>: Incorpora la rama nombrada en la actual 
    
### Paso a paso en GIT
La instalacin y puesta en marcha de GIT es muy sencilla, a continuacin se detalla un listado de pasos para instalar GIT, inicializar un repositorio, crear una rama, editar un archivo, generar un commit y finalmente incorporar los cambios en la rama principal.

1. Instalar GIT: La instalacin es muy sencilla, se debe ingresar a el sitio de GIT en el siguiente [link](https://git-scm.com/downloads) y descargar el instalador que procede para el sistema operativo propio o hacerlo mediante las herramientas de lnea de comandos respectivas:
    * Winget para windows
    * Brew para MacOS
    * Apt-get para Debian/Ubuntu
    
2. Comprobar la versin de GIT: Una vez se ha instalado git debemos asegurarnos que todo est bien y podemos hacerlo comprobando la versin en la lnea de comandos con el comando ``` git --version```

3. Inicializar el repositorio en el directorio donde se encuentra nuestro proyecto con el comando ```git init```, esto nos genera un archivo .git el cual contiene la informacin de el repositorio creado y es necesario para el trabajo de la herramienta
